```python
data = pd.read_csv("C:/Users/HP/Documents/canteen_shop_data.csv")
data.head()
```

```python
burger = data[(data['Item']=='Burger')]
drink = data[(data['Item']=='Drink')]
salad = data[(data['Item']=='Salad')]
sandwich = data[(data['Item']=='Sandwich')]
salad
```

```python
items = salad.groupby('Payment Method').sum()
items
```

```python
x = items.Total
y = items.Total.index

x
```

```python

plt.bar(y,x, color = 'pink')
plt.xticks(rotation = 30 )
plt.ylabel('Total')
plt.xlabel('Payment Method')
plt.title('SALAD')
plt.show()
```
